```python
import time as t
```

```python
vec1 = np.random.rand(100000)
vec2 = np.random.rand(100000)
```

```python
tic = t.time()
dotpro = np.dot(vec1,vec2)
toc = t.time()
print(f"Dot product is {dotpro:.2f},  and time taken is {1000* (toc-tic):.1f} ms")
```
