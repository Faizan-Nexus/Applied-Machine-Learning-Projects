```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
num = np.random.rand(10)
```

```python
style.use('ggplot')
plt.plot(num, 'g', linewidth = 2, linestyle = '-',label = 'Random Line')
# plt.plot(num, 'g', linewidth = 2, linestyle = '-.' ,label = 'Random Line')
# plt.plot(num, 'g', linewidth = 2, linestyle = ':',label = 'Random Line')
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('Thanda Plot')
plt.grid()
```

```python
plt.legend()
plt.show()
```

```python
amount = [737.63 ,737.63 ,897.4 ,1707 ,5.43 ,3447.69 ,3423.57 ,3506.19 ,3403.7 ,3729.32 ,1476 ,727.25 ,1101.9 ,1101.62]
reach = [102962 ,180175 ,173118 ,18355415 ,2448 ,1136229 ,682351 ,1077778 ,934611 ,999159 ,142743 ,114896 ,64594 ,169228]
print(f"{np.max(reach):,}")
```

```python
style.use('ggplot')
plt.plot(amount, reach, 'g', linewidth = 1.2, linestyle = '--',label = 'AD Success')
plt.axis([0,500,0,1e7])
# plt.plot(num, 'g', linewidth = 2, linestyle = '-.' ,label = 'Random Line')
# plt.plot(num, 'g', linewidth = 2, linestyle = ':',label = 'Random Line')
plt.xlabel('Amount Spent')
plt.ylabel('Reach of Ad')
plt.title('Ad Success as per Value')
plt.annotate('Max', ha = 'center', va = 'top', xytext= (500,0.2e7), xy = (500,0.53e7), arrowprops = {'facecolor' : 'green'})
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,4))
plt.subplots_adjust(hspace = .26)
plt.subplot(2,1,1)
plt.plot(amount,'g', linewidth = 1.2, linestyle = '--',label = 'AD Success')
plt.title('Amount Spent')
plt.subplot(2,1,2)
plt.title('Reach of Ad')
plt.plot(reach, 'b', linewidth = 1.2, linestyle = '-',label = "Reach")
# plt.annotate('Max', ha = 'center', va = 'top', xytext= (500,0.2e7), xy = (500,0.53e7), arrowprops = {'facecolor' : 'green'})
plt.grid()
plt.legend()
plt.show()
```
