```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bojack.jpeg')
img = plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
imgG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imghsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
fig ,axes = plt.subplots(1,3,figsize=(15,5))
axes[0].imshow(cv2.cvtColor(cv2.imread('bojack.jpeg'), cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(imgG, cmap='gray')
axes[1].set_title('Gray Image')
axes[1].axis('off')
axes[2].imshow(imghsv)
axes[2].set_title('HSV Image')
axes[2].axis('off')
```

```python
import numpy as np

img = cv2.imread("Shapes.png") 

height, width, _ = img.shape

slice_height = height // 2
slice_width = width // 2

slices = [
    img[:slice_height, :slice_width],  
    img[:slice_height, slice_width:],  
    img[slice_height:, :slice_width],  
    img[slice_height:, slice_width:]
]


fig ,axes = plt.subplots(1,5,figsize=(9,3))
axes[0].imshow(slices[0])
axes[0].set_title('Top Left')
axes[0].axis('off')
axes[1].imshow(slices[1])
axes[1].set_title('Top Right')
axes[1].axis('off')
axes[2].imshow(slices[2])
axes[2].set_title('Bottom Left')
axes[2].axis('off')
axes[3].imshow(slices[3])
axes[3].set_title('Bottom Right')
axes[3].axis('off')
axes[4].imshow(img)
axes[4].set_title('Original Image')
axes[4].axis('off')
```
