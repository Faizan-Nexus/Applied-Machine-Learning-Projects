```python
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import *
import matplotlib.pyplot as plt
```

```python
data  = np.loadtxt("C:/Users/mohtr/Desktop/Notebook/houses.txt",delimiter = ',',skiprows = 1)
```

```python
X_train = data[:,:4]
y =data[:,4]
print(np.ptp(X_train, axis = 0))
#let's normalize our data
scaler = StandardScaler()
x_norm = scaler.fit_transform(X_train)
print("After Normalization (z-norm))")
print(np.ptp(x_norm, axis = 0))
```

```python
#Printting with un_norm features
fig,ax = plt.subplots(1,4, figsize =(12,4))
features = ["size","bedroom","floor","age"]
for i in range(len(ax)):
    ax[i].scatter(X_train[:,i],y[:], label = "Unnorm")
    ax[i].set_xlabel(features[i])
    ax[i].legend()
ax[0].set_ylabel("Price")
```
