# Mini Project
This project focuses on analyzing real-world data using fundamental statistical measures:  
- Mean  
- Mode  
- Standard Deviation  
- Normal Distribution  

We will also visualize the data to gain meaningful insights. 

### *Load and Clean Dataset*
To perform data analysis, we need `pandas` for data manipulation, `numpy` for mathematical operations, `scipy` for statistics, and `matplotlib/seaborn` for visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
# Load dataset (replace 'data.csv' with actual file path)
df = pd.read_csv("SOCR-HeightWeight.csv")

print(df.head())

# Check for missing values and sum them
print("\nCheck for total missing values in dataset\n",df.isnull().sum())

# Drop missing values if any
df = df.dropna()
```

### **Statistical Analysis of Data**
Use NumPy and SciPy for calculations.

```python
# Selecting a numerical column (replace 'column_name' with actual column)
data = df['Height(Inches)']

# Mean
H_mean_value = np.mean(data)

# Mode
H_mode_value = stats.mode(data)[0]

# Standard Deviation
H_std_dev = np.std(data)
print("Statistics of Height(Inches) column")
print(f"Mean: {H_mean_value:.2f}, \nMode: {H_mode_value:.2f}, \nStandard Deviation: {H_std_dev:.2f}")
```

```python
# Selecting a numerical column (replace 'column_name' with actual column)
data = df['Weight(Pounds)']

# Mean
W_mean_value = np.mean(data)

# Mode
W_mode_value = stats.mode(data)[0]

# Standard Deviation
W_std_dev = np.std(data)
print("Statistics of Weight (Pounds) column")
print(f"Mean: {W_mean_value:.2f}, \nMode: {W_mode_value:.2f}, \nStandard Deviation: {W_std_dev:.2f}")
```

### **Visualizing Data**  
#### Normal Distribution, Scatter Plot, and Histogram  
##### 1. Normal Distribution  
A **normal distribution** (bell curve) represents data where most values cluster around the mean, with fewer values appearing as you move further from the mean. It is symmetric, with the mean, median, and mode at the center.  
##### 2. Scatter Plot  
A **scatter plot** displays the relationship between two numerical variables. Each point represents an individual data observation with coordinates **(x, y)**.  
##### 3. Histogram  
A **histogram* visualizes the frequency distribution of a single numerical variable by dividing data into bins and counting occurrences.  


```python
plt.figure(figsize=(6, 4))
plt.scatter(df['Height(Inches)'], df['Weight(Pounds)'], color='blue', alpha=0.6, edgecolors='black')
plt.xlabel("Height (Inches)")
plt.ylabel("Weight (Pounds)")
plt.title("Scatter Plot: Height vs Weight")
plt.show()

```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1. Height Distribution
axes[0].hist(df['Height(Inches)'], bins=20, color='blue', edgecolor='black', alpha=0.7)
axes[0].set_xlabel("Height(Inches)")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Height Distribution")

# 2. Weight Distribution
axes[1].hist(df['Weight(Pounds)'], bins=20, color='green', edgecolor='black', alpha=0.7)
axes[1].set_xlabel("Weight(Pounds)")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Weight Distribution")

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Normal Distribution for Height
x_height = np.linspace(min(df['Height(Inches)']), max(df['Height(Inches)']), 100)
pdf_height = stats.norm.pdf(x_height, H_mean_value, H_std_dev)

axes[0].plot(x_height, pdf_height, color='blue', linewidth=2)
axes[0].set_xlabel("Height (Inches)")
axes[0].set_ylabel("Density")
axes[0].set_title("Normal Distribution: Height")

# Normal Distribution for Weight
x_weight = np.linspace(min(df['Weight(Pounds)']), max(df['Weight(Pounds)']), 100)
pdf_weight = stats.norm.pdf(x_weight, W_mean_value, W_std_dev)

axes[1].plot(x_weight, pdf_weight, color='green', linewidth=2)
axes[1].set_xlabel("Weight (Pounds)")
axes[1].set_ylabel("Density")
axes[1].set_title("Normal Distribution: Weight")

# Adjust layout and show the plot
plt.show()
```
