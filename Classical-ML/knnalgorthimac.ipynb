````python
import numpy as np
#Generate a train set with  thre features with weight in grams  and color in intensity almost of 10 samples               
weight = np.array([157, 160, 165, 170, 175, 180, 185, 190, 195, 200])
color_intensity = np.array([4,5,6,7,8,9,10,1,2,3])
x_train = np.array([weight, color_intensity]).T
print(x_train)
y = np.array(["Apple", "Orange", "Apple", "Orange", "Apple", "Orange", "Apple", "Orange", "Apple", "Orange"]).reshape(10, 1)
print(y)
````

````python
k  = int(input("Enter the value of k between 5-8 :  "))
test_sample = list(map(int, input("Enter the fruit features to be predicted (weight and color intensity, separated by space): ").split()))
print("The test sample is : ", test_sample)
````

````python
#Calculates the manhattan distance between the test sample and each sample in the training set
def manhattan_distance(test_sample, x_train):
    return np.sum(np.abs(x_train - test_sample), axis=1)
````

````python
#Now compare the manhatan distance select with the k nearest neighbors and select the most common class among them
def knn_predict(test_sample, x_train, y, k):
    distances = manhattan_distance(test_sample, x_train)
    # Get the indices of the k nearest neighbors
    k_indices = np.argsort(distances)[:k]
    # Get the labels of the k nearest neighbors
    k_nearest_labels = y[k_indices]
    # Return the most common class label among the k nearest neighbors
    return np.bincount(k_nearest_labels.flatten()).argmax()
````

````python
# Encode the string labels into integers
label_mapping = {"Apple": 0, "Orange": 1}
y_encoded = np.array([label_mapping[label[0]] for label in y])

# Now we can predict the class of the test sample using the knn_predict function
predicted_class_index = knn_predict(test_sample, x_train, y_encoded, k)

# Decode the predicted class back to the original string label
reverse_label_mapping = {v: k for k, v in label_mapping.items()}
predicted_class = reverse_label_mapping[predicted_class_index]

print("The predicted class for the test sample is:", predicted_class)
````
