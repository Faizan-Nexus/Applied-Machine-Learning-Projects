```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Dataset

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([220, 340, 430,  480,   630, 730,])
```

### Analyzing the trainging data

```python
print("Shape of training data: ",x_train.shape)
m = x_train.shape[0]
print("number of trainign examples: ",m)

plt.scatter(x_train, y_train, marker='*', c='b')
plt.title("House Price Presiction")
plt.xlabel("Size in 1000 Square feet")
plt.ylabel("Price in 1000 dollars")
plt.show()
```

### Intitalizing Weights

```python
w = 0
b = 0
```

### Computing the Model with initialized weights

```python
def compute_model(w,b,x_train):
    m = x_train.shape[0]
    f_wb =  []
    for i in range(m):
        f_wb.append(w*x_train[i] + b)

    return f_wb
```

```python
compute_model(w,b,x_train)
```

### Computing the Cost Function

```python
def compute_cost(x,y,b,w):
    m = x.shape[0]
    f_wb = np.zeros(m)
    cost = 0
    for i in range(m):
        f_wb = x[i]*w + b
        cost = (y[i] - f_wb)**2
    cost = cost/(2*m)
    return cost
```

```python
compute_cost(x_train,y_train,b,w)
```

### Compute Gradient Descent

```python
def compute_gradient(b,w,x,y):
    m =  x.shape[0]
    f_wb = np.zeros(m)
    d_w = 0
    d_b = 0
    for i in range(m):
        f_wb = x[i]*w + b
        d_w = d_w + (f_wb - y[i])*x[i]
        d_b = d_b + (f_wb - y[i])
    d_w = d_w/m
    d_b = d_b/m
    return d_w,d_b

def gradient_descent(a,b,x,y,w,compute_cost,compute_gradient,num):
    j_history = []
    p_history = []
    for i in range(num):
        d_w , d_b = compute_gradient (b,w,x,y)
        w = w - a* d_w
        b = b - a* d_b
        if i < 100000:
            j_history.append(compute_cost(x,y,b,w))
            if i% math.ceil(num/10) == 0:
                print(f"Itteration: {i:4}, Cost : {j_history[-1]:0.2e}",
                  f"d_w: {d_w: 0.3e}, d_b: {d_b: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
    return w,b
    
```

### Checking the fucnction

```python
alpha = 0.1
w_new , b_new = gradient_descent(alpha,b,x_train,y_train,w,compute_cost,compute_gradient,99999)
print(f"(w,b) found by gradient descent: ({w_new:8.4f},{b_new:8.4f})")
```

```python
results = compute_model(w_new,b_new,x_train)
results
compute_cost(x_train,y_train,b_new,w_new)
```
