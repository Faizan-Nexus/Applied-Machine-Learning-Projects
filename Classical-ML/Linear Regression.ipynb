```python
import numpy as np
import copy,math
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
print(f"The training dataset has {X_train.shape[1]} features. It's dimensions are: {X_train.shape} it's data type is: {X_train.dtype}")
print(f"The training dataset output has {y_train.shape} dimensions and it's data type is: {y_train.dtype}")
```

```python
b_init = 0
w_init = np.array([ 0, 5.3, 23.5, 15])
print(f"w_init shape: {w_init.shape}, b_init type: {type(b_init)}")
```

```python
#Compute Cost
def cost_func(x,y,w,b):
    n = x.shape[0]
    err = 0
    for i in range(n):
        err = err + ((np.dot(w,x[i]) + b) - y[i])**2
    err = err/(2*n)
    return err
```

```python
err = cost_func(X_train,y_train,w_init,b_init)
print(f"Total Deviation from the expected values: {err}")
```

```python
#Compute Gradient
def gradient(x,y,w,b):
    m,n = x.shape
    dj_dw = np.zeros((n))
    dj_db = 0
    for i in range(m):
        err = ((np.dot(w,x[i]) + b) - y[i])
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err*x[i,j]
        dj_db = dj_db + err
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw,dj_db
```

```python
temp_dj_dw, temp_dj_db = gradient(X_train,y_train,w_init,b_init)
print(f"Temporary weights: {temp_dj_dw} and bias: {temp_dj_db} that need to be updated")
```

```python
def gd(x,y,w_init,b_init,a,itt,cost,grad):
    J_history = []
    w = copy.deepcopy(w_init)
    b = b_init
    for i in range(itt):
        dj_dw, dj_db = grad(x,y,w,b)
        w = w - a*dj_dw
        b = b - a*dj_db
        J_history.append(cost(x,y,w,b))
        if i%math.ceil(itt/10) == 0:
            print("Itterations : ",i, " Cost: ",J_history[-1])
    return w,b,J_history
```

```python
initial_w = np.zeros_like(w_init)
initial_b = 0.
iterations = 100000
alpha = 0.000008
# run gradient descent 
w_final, b_final, J_hist = gd(X_train, y_train, initial_w, initial_b,alpha,iterations,cost_func,gradient)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
```
